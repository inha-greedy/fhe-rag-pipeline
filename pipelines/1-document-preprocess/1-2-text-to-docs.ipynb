```python
def split_text(text: str, chunk_size: int = 1000) -> list:
    docs = []
    index = 0
    while text:
        chunk = text[:chunk_size]
        text = text[chunk_size:]
        docs.append({'index': index, 'document': chunk})
        index += 1
        
    return docs

splitted_docs = split_text(text=text, chunk_size=200)

splitted_docs
```
